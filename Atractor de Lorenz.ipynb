```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

El atractor de Lorenz es un sistema dinámico definido por un conjunto de ecuaciones diferenciales parciales.

Las ecuaciones que gobiernan el atractor de Lorenz son:
\begin{align}
\frac{\partial x}{\partial t} &= \sigma(y-x) \\
\frac{\partial y}{\partial t} &= x(\rho - z) - y \\
\frac{\partial z}{\partial t} &= xy - \beta z
\end{align}
Con $\sigma = 10$, $\rho = 28$ y $\beta = 8/3$

```python
def myLorenz(x,s=10, r=28, b=2.667):
    return np.array([
        s*(x[1] - x[0]),
        x[0]*(r - x[2]) - x[1],
        x[0]*x[1] - b*x[2]
    ])
```

Para aproximar la solución, utilizamos pequeños pasos de tiempo, discretizando las ecuaciones:
\begin{align}
\Delta x &= \sigma(y-x)\Delta t \\
\Delta y &= [x(\rho - z) - y] \Delta t \\
\Delta z &= [xy - \beta z]\Delta t
\end{align}

Utilizamos el método de Euler para aproximar la solución para alguna condición inicial.

```python
dt = 0.01
num_steps = 3000

# Initial values
delta = (0., 0., 0.0001)
x = np.empty((3,num_steps +1 ))
x[:,0] = (1., 1., 1.)
y = np.empty((3,num_steps +1 ))
y[:,0] = x[:,0] + delta

# Calculate steps
for i in range(num_steps):
    x[:,i+1] = x[:,i] + myLorenz(x[:,i])*dt
    y[:,i+1] = y[:,i] + myLorenz(y[:,i])*dt

# Plot
fig = plt.figure()
ax = fig.gca(projection='3d')

ax.plot(x[0],x[1],x[2], lw=1.5)
ax.plot(y[0],y[1],y[2], lw=1.5)
ax.set_title("Atractor de Lorenz")

plt.show()
```

El atractor de Lorenz es un sistema dinámico caótico, i.e. es suceptible a pequeñas perturbaciones a las condiciones iniciales.
